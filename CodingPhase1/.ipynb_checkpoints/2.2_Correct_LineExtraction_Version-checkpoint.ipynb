```python
from pptx import Presentation
from pptx.enum.shapes import MSO_SHAPE_TYPE
from pptx.shapes.connector import Connector
import matplotlib.pyplot as plt
import math
```

```python
def get_arrowhead_info(shape):
    """
    Extracts the arrowhead types (if any) from a PowerPoint shape's XML structure.

    Args:
        shape: A shape object from a PowerPoint slide (pptx.shapes.shape.Shape).

    Returns:
        tuple: A pair (head_type, tail_type) where:
            - head_type (str or None): The type of the arrowhead at the start of the line.
            - tail_type (str or None): The type of the arrowhead at the end of the line.
            - Returns None for each if no arrowhead is found.
    """
    # Retrieve the XML representation of the shape
    xml_elem = shape.element

    # Get the XML namespace mappings (needed for finding elements in the XML tree)
    ns = xml_elem.nsmap

    # Find the `<a:ln>` element (line properties) in the XML structure
    ln = xml_elem.find(".//a:ln", ns)

    # Initialize arrowhead type variables (None means no arrowhead)
    head_type = None
    tail_type = None

    # Check if the line properties (`<a:ln>`) exist in the shape
    if ln is not None:
        # Locate `<a:headEnd>` (defines the arrowhead at the start of the line)
        head = ln.find("a:headEnd", ns)
        
        # Locate `<a:tailEnd>` (defines the arrowhead at the end of the line)
        tail = ln.find("a:tailEnd", ns)

        # Extract the `type` attribute from `<a:headEnd>` if it exists
        if head is not None:
            head_type = head.get("type")

        # Extract the `type` attribute from `<a:tailEnd>` if it exists
        if tail is not None:
            tail_type = tail.get("type")

    # Return the detected arrowhead types (or None if not present)
    return head_type, tail_type
```

```python
def extract_arrow_position(slide):
    for shape in slide.shapes:
        if shape.shape_type == MSO_SHAPE_TYPE.LINE:
            shape: Connector = shape
            x1, y1 = shape.begin_x, shape.begin_y
            x2, y2 = shape.end_x, shape.end_y
```

```python
def main():
    prs = Presentation("TestFiles/test.pptx")
    for i, slide in enumerate(prs.slides, start=1):
        print(f"Slide {i}:")
        for shape in slide.shapes:
            if shape.shape_type == MSO_SHAPE_TYPE.LINE:
                shape: Connector = shape
                print(
                    f"  Line from ({shape.begin_x}, {shape.begin_y}) to ({shape.end_x}, {shape.end_y})"
                )
                arrow_head, arrow_tail = get_arrowhead_info(shape)
                print(f"    Arrowhead: {arrow_head}, Arrowtail: {arrow_tail}")
                plt.plot(
                    [shape.begin_x, shape.end_x], [shape.begin_y, shape.end_y], "k-"
                )
        plt.show()


if __name__ == "__main__":
    main()
```

# Rotate

```python
def rotate_line(begin_x, begin_y, end_x, end_y, rot_degrees):
    mid_x = (begin_x + end_x) / 2
    mid_y = (begin_y + end_y) / 2

    theta = math.radians(rot_degrees)

    def rotate_point(x, y):
        translated_x = x - mid_x
        translated_y = y - mid_y

        rotated_x = translated_x * math.cos(theta) - translated_y * math.sin(theta)
        rotated_y = translated_x * math.sin(theta) + translated_y * math.cos(theta)

        return rotated_x + mid_x, rotated_y + mid_y

    new_begin_x, new_begin_y = rotate_point(begin_x, begin_y)
    new_end_x, new_end_y = rotate_point(end_x, end_y)

    return new_begin_x, new_begin_y, new_end_x, new_end_y
```

```python
def main():
    prs = Presentation("TestFiles/test.pptx")
    plt.gca().invert_yaxis()
    for i, slide in enumerate(prs.slides, start=1):
        print(f"Slide {i}:")
        for shape in slide.shapes:
            if shape.shape_type == MSO_SHAPE_TYPE.LINE:
                shape: Connector = shape
                begin_x = shape.begin_x
                begin_y = shape.begin_y
                end_x = shape.end_x
                end_y = shape.end_y
                rot = shape.element.rot
                begin_x, begin_y, end_x, end_y = rotate_line(
                    begin_x, begin_y, end_x, end_y, rot
                )
                print(f"  Line from ({begin_x}, {begin_y}) to ({end_x}, {end_y})")
                arrow_head, arrow_tail = get_arrowhead_info(shape)
                print(f"    Arrowhead: {arrow_head}, Arrowtail: {arrow_tail}")
                plt.plot([begin_x, end_x], [begin_y, end_y], "k-")
        plt.show()


if __name__ == "__main__":
    main()
```
