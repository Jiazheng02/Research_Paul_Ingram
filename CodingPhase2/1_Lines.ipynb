```python
import cv2
import matplotlib.pyplot as plt
import LinesProcessor
```

```python
image_path = "/Users/jiazhengtian/Desktop/image.png"
```

```python
lineprocess = LinesProcessor.LinesDetector(image_path)
```

```python
lines = lineprocess.find_lines()
```

```python
image = cv2.imread(image_path)

# 在图像上绘制线条（绿色，粗细2）
for (x1, y1, x2, y2) in lines:
    cv2.line(image, (x1, y1), (x2, y2), color=(0, 255, 0), thickness=2)

# OpenCV 是 BGR，matplotlib 使用 RGB，因此需要转换颜色格式
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 使用 matplotlib 显示图像
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.title("Lines on Image")
plt.axis("off")
plt.show()
```
