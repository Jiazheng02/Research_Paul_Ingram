```python
import os
from PPTXProcessor import PPTXProcessor
```

```python
id_path = "combined id_ego thru fall 24.xlsx"
paths = [
    "2021 A values.pptx",
    "2021 B values.pptx"
]
```

```python
for pptx_path in paths:

    extractor = PPTXProcessor(pptx_path, id_path)
    df = extractor.process_pptx_to_dataframe()
    
    # Remove file extension from pptx_path to avoid ".pptx.xlsx"
    output_filename = os.path.splitext(os.path.basename(pptx_path))[0] + "_extracted.xlsx"
    
    output_path = f"Extracted Data 3/{output_filename}"
    
    df.to_excel(output_path, index=False, engine="openpyxl")
```
