```python
import matplotlib.pyplot as plt
import seaborn as sns

# 
columns = ['speed_vec', 'optimism_vec', 'change_vec', 'holistic_vec', 'creative_vec']

plt.figure(figsize=(10, 6))

# KDE
for col in columns:
    sns.kdeplot(df[col], label=col, linewidth=2)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Trait Vectors')
plt.legend()
plt.grid(True)
plt.show()
```
